# EDA - 2016 hourly plot tests



```python
import numpy as np;
import pandas as pd;

import matplotlib.pyplot as plt
import seaborn as sns;



botanic_WH_2016 = pd.read_csv('../data/preprocessed/botanic_hourly_2016_unique_timestamp.csv', index_col=False);

botanic_WH_2016.head()[['OB_TIME','AIR_TEMPERATURE','RLTV_HUM','DEWPOINT','MSL_PRESSURE','STN_PRES','WIND_SPEED','WIND_DIRECTION','Q10MNT_MXGST_SPD','ALT_PRES']]

# botanic_WH_2016.shape
# botanic_WH_2016.columns.to_numpy()

botanic_WH_2016['AIR_TEMPERATURE'] = pd.to_numeric( botanic_WH_2016['AIR_TEMPERATURE'], errors='coerce')
botanic_WH_2016['RLTV_HUM'] = pd.to_numeric( botanic_WH_2016['RLTV_HUM'], errors='coerce')
botanic_WH_2016['DEWPOINT']        = pd.to_numeric( botanic_WH_2016['DEWPOINT'], errors='coerce')
botanic_WH_2016['OB_TIME']         = pd.to_datetime( botanic_WH_2016['OB_TIME'] )





gogar_WH_2016 = pd.read_csv('../data/preprocessed/gogar_hourly_2016_unique_timestamp.csv', index_col=False);

gogar_WH_2016.head()[['OB_TIME','AIR_TEMPERATURE','RLTV_HUM','DEWPOINT','MSL_PRESSURE','STN_PRES','WIND_SPEED','WIND_DIRECTION','Q10MNT_MXGST_SPD','ALT_PRES']]

gogar_WH_2016['OB_TIME']         = pd.to_datetime( gogar_WH_2016['OB_TIME'] )
gogar_WH_2016['AIR_TEMPERATURE'] = pd.to_numeric( gogar_WH_2016['AIR_TEMPERATURE'], errors='coerce')
gogar_WH_2016['RLTV_HUM']        = pd.to_numeric( gogar_WH_2016['RLTV_HUM'], errors='coerce')
gogar_WH_2016['DEWPOINT']        = pd.to_numeric( gogar_WH_2016['DEWPOINT'], errors='coerce')
gogar_WH_2016['MSL_PRESSURE']    = pd.to_numeric( gogar_WH_2016['MSL_PRESSURE'], errors='coerce')
gogar_WH_2016['STN_PRES']        = pd.to_numeric( gogar_WH_2016['STN_PRES'], errors='coerce')
gogar_WH_2016['WIND_SPEED']        = pd.to_numeric( gogar_WH_2016['WIND_SPEED'], errors='coerce')
gogar_WH_2016['WIND_DIRECTION']        = pd.to_numeric( gogar_WH_2016['WIND_DIRECTION'], errors='coerce')






plt.figure( figsize=(15,6) )
ax = sns.lineplot( x=botanic_WH_2016['OB_TIME'], y=botanic_WH_2016['AIR_TEMPERATURE'], label='Botanic Garden');
ax = sns.lineplot( ax=ax, x=gogar_WH_2016['OB_TIME'], y=gogar_WH_2016['AIR_TEMPERATURE'], label='Gogar Bank');
plt.title('Temperature at Edinburgh in 2016')
plt.xlabel('Date')
plt.ylabel('Air Temperature ( Celsius )')
plt.grid()

ax.legend()

plt.show()



plt.figure( figsize=(15,6) )
ax = sns.lineplot( x=botanic_WH_2016['OB_TIME'], y=botanic_WH_2016['DEWPOINT'], label='Botanic Garden');
ax = sns.lineplot( ax=ax, x=gogar_WH_2016['OB_TIME'], y=gogar_WH_2016['DEWPOINT'], label='Gogar Bank');
plt.title('Dewpoint Temperature ( temp. stat. more connected to human comfort ) at Edinburgh in 2016')
plt.xlabel('Date')
plt.ylabel(' Dewpoint Temperature ( Celsius )')
plt.grid()

ax.legend()

plt.show()



plt.figure( figsize=(15,6) )
ax = sns.lineplot( x=botanic_WH_2016['OB_TIME'], y=botanic_WH_2016['RLTV_HUM'], label='Botanic Garden');
ax = sns.lineplot( ax=ax, x=gogar_WH_2016['OB_TIME'], y=gogar_WH_2016['RLTV_HUM'], label='Gogar Bank');
plt.title('Relative Humidity at Edinburgh in 2016')
plt.xlabel('Date')
plt.ylabel('Relative Humidity ( Percentage )')
plt.grid()

ax.legend()

plt.show()


plt.figure( figsize=(15,6) )
ax = sns.lineplot( x=gogar_WH_2016['OB_TIME'], y=gogar_WH_2016['MSL_PRESSURE'], label='Mean sea level air pressure', color='green');
ax = sns.lineplot( ax=ax, x=gogar_WH_2016['OB_TIME'], y=gogar_WH_2016['STN_PRES'], label='Station air pressure', color='pink');
plt.title('Atmospheric pressure at Gogar Bank in 2016')
plt.xlabel('Date')
plt.ylabel('Air pressure ( hpa )')
plt.grid()

ax.legend()

plt.show()


plt.figure( figsize=(15,6) )
ax = sns.lineplot( x=gogar_WH_2016['OB_TIME'], y=gogar_WH_2016['WIND_SPEED'], label='Wind Speed', color='green');
#ax = sns.lineplot( ax=ax, x=gogar_WH_2016['OB_TIME'], y=gogar_WH_2016['WIND_DIRECTION'], label='Wind Direction', color='pink');
plt.title('Wind Information at Gogar Bank in 2016')
plt.xlabel('Date')
plt.ylabel('Wind Speed ( knots )')
plt.grid()

ax.legend()

plt.show()

```
